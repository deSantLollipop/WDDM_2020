```python
lista =[]
with open('./Top100.txt', 'r') as f:
    for linia in f:
        lista.append(linia)

str = ""

lista1 = lista[1:]

lil_1=lista1[0::3]

lil_2 = []
for n in range(len(lista1)//3):
    lil_2.append(' '.join(lista1[n*3+1:n*3+3]))

val_1 = []
val_2 = []
val_3 = []
val_4 = []
val_5 = []
val_6 = []
val_7 = []

for i in lil_1:
        str = i
        strx=str.replace("\n","").split("\t")
        val_1.append(strx[0])
        val_2.append(strx[1])

strx = []
num=0
tab="   "
for i in lil_2:
        str = i
        strx=str.replace("\n","").split("\t",2)
        val_3.append(strx[1])
        val_2[num] = val_2[num] + tab + strx[0]
        num = num + 1
        str = strx[2] + ' 0' 
        str = str.replace(",","")
        strx=str.split()
        val_4.append(strx[0])
        val_5.append(strx[1])
        val_6.append(strx[2])
        val_7.append(strx[3])  


from itertools import repeat
values = [[] for i in repeat(None, 100)]

for i in range(100):
    values[i] = [val_1[i],val_2[i],val_3[i],val_4[i],val_5[i],val_6[i],val_7[i]]


bigdic  = []
lildic = {}
keys = ['Rank' , 'Site' , 'System' , 'Cores' , 'Rmax (TFlop/s)' , 'Rpeak (TFlop/s)', 'Power (kW)']

for i in values:
    lildic = dict(zip(keys, i))
    bigdic.append(lildic)
```
